```python
import pandas as pd
import calplot
import numpy as np; np.random.seed(sum(map(ord, 'calplot')))
import matplotlib.pyplot as plt
from pathlib import Path
```

```python
#read my csv and make DF
#2018
path2018 = Path("Resources/daily_aqi_by_county_2018.csv")
data2018 = pd.read_csv(path2018)
aqi2018_df = pd.DataFrame(data2018)
#aqi2018_df.head()

#2019
path2019 = Path("Resources/daily_aqi_by_county_2019.csv")
data2019 = pd.read_csv(path2019)
aqi2019_df = pd.DataFrame(data2019)
#aqi2019_df.head()

#2020
path2020 = Path("Resources/daily_aqi_by_county_2020.csv")
data2020 = pd.read_csv(path2020)
aqi2020_df = pd.DataFrame(data2020)
#aqi2020_df.head()

#2021
path2021 = Path("Resources/daily_aqi_by_county_2021.csv")
data2021 = pd.read_csv(path2021)
aqi2021_df = pd.DataFrame(data2021)
#aqi2021_df.head()

#2022
path2022 = Path("Resources/daily_aqi_by_county_2022.csv")
data2022 = pd.read_csv(path2022)
aqi2022_df = pd.DataFrame(data2022)
#aqi2022_df.head()

#2023
path2023 = Path("Resources/daily_aqi_by_county_2023.csv")
data2023 = pd.read_csv(path2023)
aqi2023_df = pd.DataFrame(data2023)
aqi2023_df.head()
```

```python
#only nj data
#2018
aqi2018nj_df = aqi2018_df[aqi2018_df['State Name'] == 'New Jersey']
#aqi2018nj_df.head()

#2019
aqi2019nj_df = aqi2019_df[aqi2019_df['State Name'] == 'New Jersey']
#aqi2019nj_df.head()

#2020
aqi2020nj_df = aqi2020_df[aqi2020_df['State Name'] == 'New Jersey']
#aqi2020nj_df.head()

#2021
aqi2021nj_df = aqi2021_df[aqi2021_df['State Name'] == 'New Jersey']
#aqi2021nj_df.head()

#2022
aqi2022nj_df = aqi2022_df[aqi2020_df['State Name'] == 'New Jersey']
#aqi2022nj_df.head()

#2023
aqi2023nj_df = aqi2023_df[aqi2023_df['State Name'] == 'New Jersey']
aqi2023nj_df.head()
```

```python
#clean for pie chart
##counts for 2018 nj
days2018_count = aqi2018nj_df["Category"].value_counts()
#print(days2018_count)

#counts for 2019
days2019_count = aqi2019nj_df["Category"].value_counts()
#print(days2019_count)

#counts for 2020
days2020_count = aqi2020nj_df["Category"].value_counts()
#print(days2020_count)

#counts for 2021
days2021_count = aqi2021nj_df["Category"].value_counts()
#print(days2021_count)

#counts for 2022
days2022_count = aqi2022nj_df["Category"].value_counts()
#print(days2022_count)

#counts for 2023
days2023_count = aqi2023nj_df["Category"].value_counts()
print(days2023_count)

# Replace NaN values with 0 in each of the Series or DataFrames
days2018_count = days2018_count.fillna(0)
days2019_count = days2019_count.fillna(0)
days2020_count = days2020_count.fillna(0)
days2021_count = days2021_count.fillna(0)
days2022_count = days2022_count.fillna(0)
days2023_count = days2023_count.fillna(0)
```

```python
#counts for 2018 to 2023
# Concatenate the DataFrames
#combined_counts = pd.concat([days2018_count, days2019_count, days2020_count, days2021_count, days2022_count, days2023_count], axis=1)
#print(combined_counts)
days18to23=(days2018_count)+(days2019_count)+(days2020_count)+(days2021_count)+(days2022_count)+(days2023_count)
days18to23 = days18to23.fillna(0)
print(days18to23)
```

```python
# Generate a pie chart for day category 2018-2023
labels = ['Good','Moderate','Unhealthy','Unhealthy for Sensitive Groups','Very Unhealthy','Hazardous']
colors = ['lightgreen','orange','lightpink','red','coral','navy']
explode_all = (0, 0.1, 0, 0, 0.4, 0.7) 
#days18to23.plot(kind='pie', labels=labels, colors =colors, legend = True, autopct='%2.2f%%')
#plt.pie(days18to23,labels=labels,colors=colors,autopct='%2.2f%%')
plt.pie(days18to23,labels=None,colors=colors,autopct='%2.2f%%', pctdistance=1.20, explode=explode_all)
plt.title('Category of Days 2018-2023')
# Move the legend to a specific position
plt.legend(labels=labels,loc='upper right', bbox_to_anchor=(1.75, 1))
plt.show()

#pie chart 2018
explode_2018=(0, 0, 0, 0, 0.4, 0.8)
plt.pie(days2018_count,labels=None,colors=colors,autopct='%2.2f%%', pctdistance=1.20, explode=explode_2018)
plt.title('Category of Days 2018')
# Move the legend to a specific position
plt.legend(labels=labels,loc='upper right', bbox_to_anchor=(1.75, 1))
plt.show()

#pie chart 2019
explode_2019=(0, 0, 0, 0.4, 0.7, 1)
plt.pie(days2019_count,labels=None,colors=colors,autopct='%2.2f%%', pctdistance=1.20, explode=explode_2019)
plt.title('Category of Days 2019')
# Move the legend to a specific position
plt.legend(labels=labels,loc='upper right', bbox_to_anchor=(1.75, 1))
plt.show()

#pie chart 2020
explode_2020=(0, 0, 0, 0.4, 0.7, 1)
plt.pie(days2020_count,labels=None,colors=colors,autopct='%2.2f%%', pctdistance=1.20, explode=explode_2020)
plt.title('Category of Days 2020')
# Move the legend to a specific position
plt.legend(labels=labels,loc='upper right', bbox_to_anchor=(1.75, 1))
plt.show()

#pie chart 2021
explode_2021=(0, 0, 0, 0.4, 0.7, 1)
plt.pie(days2021_count,labels=None,colors=colors,autopct='%2.2f%%', pctdistance=1.20, explode=explode_2021)
plt.title('Category of Days 2021')
# Move the legend to a specific position
plt.legend(labels=labels,loc='upper right', bbox_to_anchor=(1.75, 1))
plt.show()

#pie chart 2022
explode_2022=(0, 0, 0, 0.4, 0.7, 1)
plt.pie(days2022_count,labels=None,colors=colors,autopct='%2.2f%%', pctdistance=1.20, explode=explode_2022)
plt.title('Category of Days 2022')
# Move the legend to a specific position
plt.legend(labels=labels,loc='upper right', bbox_to_anchor=(1.75, 1))
plt.show()

#pie chart 2023
explode_2023=(0, 0, 0, 0.4, 0.7, 1)
plt.pie(days2023_count,labels=None,colors=colors,autopct='%2.2f%%', pctdistance=1.20, explode=explode_2023)
plt.title('Category of Days 2023')
# Move the legend to a specific position
plt.legend(labels=labels,loc='upper right', bbox_to_anchor=(1.75, 1))
plt.show()
```
