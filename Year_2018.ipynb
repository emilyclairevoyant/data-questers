```python
import pandas as pd
from pathlib import Path
import matplotlib.pyplot as plt
```

```python
# Save file path to variable
aqi_2018 = Path("Resources/daily_aqi_by_county_2018.csv")

# Read with Pandas
aqi_2018_df= pd.read_csv(aqi_2018)
aqi_2018_df.head()
```

```python
#Groupby State
nj_2018_df = aqi_2018_df.loc[aqi_2018_df['State Name'] == 'New Jersey']


nj_2018_df
```

```python
# Find the mean AQI values
#nj_aqi_2018_group= nj_2018_df.groupby(["county Name", "AQI"])

Mean_AQI = nj_2018_df.groupby("county Name").mean(numeric_only=True)["AQI"]
Mean_AQI
```

```python
# Convert into a dataframe
AQI_2018 = pd.DataFrame({"Mean AQI": Mean_AQI})

AQI_2018
```

```python
# Generate a bar plot showing how the Mean AQI Values are changing for counties using pyplot.
x_axis = AQI_2018.index.values
y_axis = AQI_2018["Mean AQI"]

plt.bar(x_axis, y_axis, color='tab:blue', alpha=0.8, align='center')

plt.title("County vs. Mean AQI")
plt.xlabel("County Names")
plt.ylabel("Mean AQI")
plt.xticks(rotation="vertical")

plt.show()
```

```python
#Groupby Defining parameter PM2.5
dfparam_PM25_df = nj_2018_df.loc[nj_2018_df['Defining Parameter'] == 'PM2.5']
dfparam_PM25_df

# Find the mean AQI values
aqi_dfparam_PM25 = dfparam_PM25_df.groupby("county Name").mean(numeric_only=True)["AQI"]
aqi_dfparam_PM25
```

```python
#Groupby Defining parameter Ozone
dfparam_ozone_df = nj_2018_df.loc[nj_2018_df['Defining Parameter'] == 'Ozone']
dfparam_ozone_df

# Find the mean AQI values
aqi_dfparam_ozone = dfparam_ozone_df.groupby("county Name").mean(numeric_only=True)["AQI"]
aqi_dfparam_ozone
```

```python
#Groupby Defining parameter NO2
dfparam_NO2_df = nj_2018_df.loc[nj_2018_df['Defining Parameter'] == 'NO2']
dfparam_NO2_df

# Find the mean AQI values
aqi_dfparam_NO2 = dfparam_NO2_df.groupby("county Name").mean(numeric_only=True)["AQI"]
aqi_dfparam_NO2
```

```python
#Groupby Defining parameter CO
dfparam_CO_df = nj_2018_df.loc[nj_2018_df['Defining Parameter'] == 'CO']
dfparam_CO_df

# Find the mean AQI values
aqi_dfparam_CO = dfparam_CO_df.groupby("county Name").mean(numeric_only=True)["AQI"]
aqi_dfparam_CO
```

```python
# convert in to a dataframe with summary table 
summary_2018_df= pd.DataFrame({"Mean AQI": Mean_AQI,
                         "Mean AQI By PM2.5": aqi_dfparam_PM25,
                         "Mean AQI By Ozone": aqi_dfparam_ozone,
                         "Mean AQI By NO2": aqi_dfparam_NO2,
                         "Mean AQI By CO": aqi_dfparam_CO})
summary_2018_df
```

```python
# Generate a bar plot showing how the Mean AQI By PM2.5 Values are changing for counties using pyplot.
x_axis = summary_2018_df.index.values
y_axis = summary_2018_df["Mean AQI By PM2.5"]

plt.bar(x_axis, y_axis, color='grey', alpha=0.8, align='center')

plt.title("County vs. Mean AQI By PM2.5")
plt.xlabel("County Names")
plt.ylabel("Mean AQI By PM2.5")
plt.xticks(rotation="vertical")

plt.show()
```

```python
# Generate a bar plot showing how the Mean AQI By Ozone Values are changing for counties using pyplot.
x_axis = summary_2018_df.index.values
y_axis = summary_2018_df["Mean AQI By Ozone"]

plt.bar(x_axis, y_axis, color='pink', alpha=0.8, align='center')

plt.title("County vs. Mean AQI By Ozone")
plt.xlabel("County Names")
plt.ylabel("Mean AQI By Ozone")
plt.xticks(rotation="vertical")

plt.show()
```

```python
# Generate a bar plot showing how the Mean AQI By NO2 Values are changing for counties using pyplot.
x_axis = summary_2018_df.index.values
y_axis = summary_2018_df["Mean AQI By NO2"]

plt.bar(x_axis, y_axis, color='lightblue', alpha=0.8, align='center')

plt.title("County vs. Mean AQI By NO2")
plt.xlabel("County Names")
plt.ylabel("Mean AQI By NO2")
plt.xticks(rotation="vertical")

plt.show()
```

```python
# Generate a bar plot showing how the Mean AQI By NO2 Values are changing for counties using pyplot.
x_axis = summary_2018_df.index.values
y_axis = summary_2018_df["Mean AQI By CO"]

plt.bar(x_axis, y_axis, color='red', alpha=0.8, align='center')

plt.title("County vs. Mean AQI By CO")
plt.xlabel("County Names")
plt.ylabel("Mean AQI By CO")
plt.xticks(rotation="vertical")

plt.show()
```
